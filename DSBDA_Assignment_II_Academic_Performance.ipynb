### Scanning and Removing Outliers

**Scanning and removing outliers from raisedhands using IQR technique**

```python
datafill['raisedhands'].describe()
```

```python
# We can directly take 25% and 75% values from describe() function
# Q1 = 13.750000
# Q3 = 74.250000
Q1 = datafill['raisedhands'].quantile(0.25)
Q3 = datafill['raisedhands'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - (1.5 * IQR)
upper = Q3 + (1.5 * IQR)
```

Scanning for Outliers in raisedhands

```python
# checking if there exists any outliers
datafill[(datafill['raisedhands'] < lower) | (datafill['raisedhands'] > upper)]
```
