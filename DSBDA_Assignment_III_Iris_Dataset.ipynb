Write a Python program to display some basic statistical details

like percentile, mean, standard deviation etc. of the species of ‘Iris-
setosa’, ‘Iris-versicolor’ and ‘Iris-virginica’ of iris.csv dataset.

Calculate the measures of variability. Calculate and provide the
visualization of the Correlation among the variables.

```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('/content/Iris.csv')
```

```python
dataframe.head()
```

Basic Statistical Data for  ‘Iris- setosa’, ‘Iris-versicolor’ and ‘Iris-virginica’

```python
# mean
groupBySpecies = dataframe.groupby('Species')
groupBySpecies.mean()
```

```python
# median
groupBySpecies.median()
```

```python
# variance
groupBySpecies.var()
```

```python
# percentile
groupBySpecies.quantile()
```

```python
# standard deviation
groupBySpecies.std()
```

Calculating Correlation between variables

```python
dataframe.corr()
```

Visualizing Correlation using heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(dataframe.corr())
plt.show()
```
